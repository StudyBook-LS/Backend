```python
STOP_WORDS = nlp.Defaults.stop_words.union(['\n', '\n\n', '=', '>', 'i', 'cc:', 'to:',
                                            'from', 'subject: ', 'the', 'to: ', 'am',
                                           ' ', '  ', 'subject:'])
```

```python
#nlp = spacy.load("en_core_web_sm")
nlp = spacy.load("en")


# Tokenizer
tokenizer = Tokenizer(nlp.vocab)
```

```python
tokens = []

for doc in tokenizer.pipe(df['brief_summary'], batch_size=500):
    
    doc_tokens = []
    
    for token in doc: 
        if token.text not in STOP_WORDS:
            #doc_tokens.append(token.text.lower().strip() if token != '')
            token = token.text.lower().strip()
            if token != '':
                doc_tokens.append(token)
   
    tokens.append(doc_tokens)

df['tokens'] = tokens
```
