## Function for test files

This function will create a dataframe containing xml files that are located in a single directory

```python
def generate_dataframe(filefolder, columns):
    
    df = pd.DataFrame(columns=columns)
    
    files = os.listdir(filefolder)
    
    for xml in files:
        print(xml)
        
        path = os.path.join(filefolder, xml)
        
        if  path[-3:] == 'xml': 
            with open(path, 'rb') as f2:
                data = f2.read()
                soup = BeautifulSoup(data, 'xml')

                data_dict = test_dict(soup)
                temp_df = pd.DataFrame.from_dict([data_dict])
                df = pd.concat([df, temp_df])
            
                
    return df
```

## Function for Paul

Run this function to find all the unique tags for the totality of the data downloaded. <br>
Pass in the root folder

```python
def generate_tag_set(filefolder):
    
    clinic_list = {'address'}

    
    folder = os.listdir(filefolder)
    
    for files in folder:
        for xml in files: 

            path = os.path.join(filefolder, xml)
            print(path)

            with open(path, 'rb') as f2:
                data = f2.read()
                soup = BeautifulSoup(data, 'xml')
                list_of_tags = [tag.name for tag in soup.find_all()]
                for i in list_of_tags:
                    clinic_list.add(i)

    return clinic_list
```
